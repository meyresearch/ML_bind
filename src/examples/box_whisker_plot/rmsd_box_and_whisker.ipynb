```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
raw_rmsd= pd.read_csv("rmsd_results.csv")
```

```python
sns.set_theme(style="whitegrid")
```

```python
plt.figure(figsize=(30,15))
plt.rc('xtick',labelsize=24)
plt.rc('ytick',labelsize=28)
plt.xlabel("Ligand", size=32, labelpad=20)
plt.ylabel("RMSD (nm)", size= 32, labelpad=20)
rmsd=sns.boxplot(x=raw_rmsd["Ligand"], y=raw_rmsd["RMSD (nm)"], width=0.75, fliersize=10, linewidth=4)
plt.savefig("rmsd_box_whisker.png")
```

```python
plt.figure(figsize=(30,15))
plt.rc('xtick',labelsize=24)
plt.rc('ytick',labelsize=28)
plt.xlabel("Ligand", size=32, labelpad=20)
plt.ylabel("RMSD (nm)", size= 32, labelpad=20)
rmsd=sns.boxplot(x=raw_rmsd["Ligand"], y=raw_rmsd["RMSD (nm)"], width=0.75, fliersize=10, linewidth=4, color="steelblue")
plt.savefig("rmsd_box_whisker_blue.png")
```
